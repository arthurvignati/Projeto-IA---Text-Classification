# Classificação de E-mails Spam vs Ham

## Objetivo
Desenvolver um sistema de inteligência artificial capaz de identificar automaticamente e-mails **spam** (não desejados) e **ham** (legítimos).  
O projeto será estruturado em três etapas:
1. Criação de baselines com algoritmos clássicos de NLP.
2. Evolução para modelos intermediários com embeddings.
3. Ajuste fino com LLMs modernos para maximizar desempenho.

## Fontes de Dados
- **Enron Email Dataset**  
  Conjunto de dados público de e-mails trocados entre funcionários da Enron Corporation. Será de grande utilidade, pois essa base de dados mostra realmente como as pessoas se comunicavam na empresa, identificando hierarquias e fluxos de informação, de tal forma que poderemos desenvolver   e testar algoritmos para identificar comportamentos suspeitos e atividades fraudulentas, ou até mesmo validar se aquele email é realmente legítimo.

- **SpamAssassin Public Corpus**  
  Coleção pública de e-mails rotulados como spam e ham, que servirá de grande ajuda para montar a baseline e o treinamento da inteligência, para desenvolver e testar modelos de detecção de spam.  

## Ferramentas e Técnicas

  O projeto será desenvolvido com foco em técnicas mais acessíveis de **processamento de texto** e **aprendizado de máquina**, para que possamos trabalhar dentro do escopo de conhecimento já adquirido na disciplina.  

  O primeiro passo será o **pré-processamento dos e-mails**, onde aplicaremos etapas simples de limpeza de texto, como a remoção de caracteres especiais, normalização de letras (tudo para minúsculo) e separação das palavras em tokens. Em seguida, vamos transformar os textos em representações numéricas usando métodos básicos, como **contagem de palavras (Bag of Words)** ou **TF-IDF**, que permitem calcular a importância de cada termo dentro da mensagem.  

  Com os dados já vetorizados, treinaremos modelos de classificação clássicos que são fáceis de implementar e compreender, como o **Naive Bayes** (amplamente utilizado em filtros de spam) e a **Regressão Logística** (que se conecta com o conteúdo que já estudamos em sala). Esses modelos servirão como nossa baseline inicial.  

  A avaliação do sistema será feita com métricas simples e interpretáveis, como **acurácia** (percentual de acertos) e **matriz de confusão** (para verificar a diferença entre falsos positivos e falsos negativos). Dessa forma, conseguiremos medir o desempenho do modelo e verificar onde ele precisa ser ajustado.  

##Resultado Esperados

Desenvolvimento de uma interface simples na qual o usuário poderá colar o corpo do e-mail como input, e receberá um output com a classificação daquele conteúdo (Spam/Ham), junto com a justificativa (pensamos em atribuir a palavras-chaves que caracterizam se é legítimo ou não).
